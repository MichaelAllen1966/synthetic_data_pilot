## Code/Results

### Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from datetime import date

# Turn warnings off for notebook publication
import warnings
warnings.filterwarnings("ignore")
```

### Define Functions

```python
def scatterplot(data, filename):
    """
    Plot and save data
    
    Inputs
    ------
    data: Data to plot as a matrixed scatter plot
    filename: filename to save (without directory or extension)
    
    Returns
    -------
    plot (and saves files)
    
    """
    # Create date string to prefix filename
    d = date.today()
    date_str = d.strftime("%y%m%d")
    
    sns.set(style="ticks")
    spm= sns.pairplot(data,hue="malignant")
    filename = './Output/' + date_str + '_' + filename + '.png'
    spm.savefig(filename)
                
    return spm
```

### Import data

```python
# Load data and drop 'id' column
original_data = pd.read_csv('./wisconsin.csv')
original_data.drop('id', axis=1, inplace=True)

# Change 'diagnosis' column to 'malignant', and put in last column place
malignant = pd.DataFrame()
original_data['malignant'] = original_data['diagnosis'] == 'M'
original_data.drop('diagnosis', axis=1, inplace=True)
```

```python
original_data.head()
```

### Plot data 

The following plots are for mean, se and worst values in the orginal data set

```python
# Divide data for plots

column_names =list(original_data)

data_features_1 = pd.DataFrame(original_data, columns = \
                     [s for s in column_names if "_mean" in s]+["malignant"])

data_features_2 = pd.DataFrame(original_data, columns = \
                     [s for s in column_names if "_se" in s]+["malignant"])

data_features_3 = pd.DataFrame(original_data, columns = \
                     [s for s in column_names if "_worst" in s]+["malignant"])
```

Scatter plot of 'mean' column values

```python
scatterplot(data_features_1, 'scatterplot_mean');
```

Scatter plot of 'se' column values

```python
scatterplot(data_features_2, 'scatterplot_sem');
```

Scatter plot of 'worst' column values

```python
scatterplot(data_features_3, 'scatterplot_worst');
```

### Descriptive statistics

```python
# Split benign and malignant DataFrames
malignant_df = original_data[original_data['malignant']==True]
benign_df = original_data[original_data['malignant']==False]
```

```python
# Get means and summarise in a bew DataFrame
description = pd.DataFrame()
description['m_mean'] = malignant_df.mean()
description['b_mean'] = benign_df.mean()
description['m/b ratio'] = description['m_mean'] / description['b_mean'] 

# Show table
description
```

## Conclusion

Scatter plots and descriptive statistics show promise that malignant and benign samples will be accessible to separation by machine learning methods.
